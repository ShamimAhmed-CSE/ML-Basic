Linear regression with a single variable, often referred to as simple linear regression, is a basic statistical method used for modeling the relationship between two variables: one independent variable (predictor) and one dependent variable (outcome). The goal is to find a linear equation that best describes this relationship.

Here's the general equation for simple linear regression:


**y=mx+b**

Where:

1. y is the dependent variable (the one you want to predict).
2. x is the independent variable (the one you're using to make predictions).
3. m is the slope of the line (the coefficient that represents the relationship between x and y).
4. b is the y-intercept (the point where the line crosses the y-axis).

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/canada-per-capita-income/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df.year,df.per_capita_income,color='red',marker='*')
```

```python
new_df = df.drop('per_capita_income',axis='columns')
new_df
```

```python
per_capita_income = df.per_capita_income
per_capita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```
