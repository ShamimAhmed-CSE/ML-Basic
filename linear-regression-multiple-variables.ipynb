Linear regression with multiple variables is also known as multiple linear regression. It's an extension of simple linear regression, where we have more than one independent variable that we use to predict a dependent variable. In multiple linear regression, we are trying to fit a linear equation to the data with the form:

**Y=b + m1*x1 + m2*x2 + m3*x3 ........**

Y is the dependent variable we are trying to predict.
X1,x2,x3 ...... are the independent variables or features.
b is the intercept (the value of Y when all independent variables are zero).
m1,m2,m3.....are the coefficients that represent the change in Y for a one-unit change in the corresponding X while holding all other variables constant.

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/homeprices/homeprices.csv')
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
reg.coef_
```

```python
reg.predict([[2500, 4, 5]])
```
